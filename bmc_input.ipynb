```python
import numpy as np
from tensorflow.keras.models import load_model

# load trained model
model = load_model("signal_classifier.h5")

print("Model loaded successfully ")
```

```python
classes = [
    'OOK','4ASK','8ASK','BPSK','QPSK','8PSK',
    '16PSK','32PSK','16APSK','32APSK','64APSK','128APSK',
    '16QAM','32QAM','64QAM','128QAM','256QAM',
    'AM-SSB-WC','AM-SSB-SC','AM-DSB-WC','AM-DSB-SC',
    'FM','GMSK','OQPSK'
]

print("Classes loaded ")
```

```python
import numpy as np
import os

def get_input_signal():
    file_path = input("Enter path of .npy signal file: ")
    
    if not os.path.exists(file_path):
        print("File not found")
        return None
    
    try:
        signal = np.load(file_path)
        
        # check shape
        if signal.shape != (2, 1024):
            print(" Invalid shape. Expected (2,1024)")
            return None
        
        return signal
    
    except Exception as e:
        print(" Error loading file:", e)
        return None
```

```python
def predict_signal(model, signal, classes):
    import numpy as np

    # preprocess
    signal = signal.T
    signal = signal / np.max(np.abs(signal))
    signal = np.expand_dims(signal, axis=-1)   # (2,1024,1)
    signal = np.expand_dims(signal, axis=0)    # (1,2,1024,1)
    
    # prediction
    pred = model.predict(signal)

    # get class
    pred_class = np.argmax(pred)
    confidence = np.max(pred)

    return classes[pred_class], confidence
```

```python
import h5py
import numpy as np

file = h5py.File("GOLD_XYZ_OSC.0001_1024.hdf5", "r")

X = file['X']
Y = file['Y']

indices = np.random.choice(X.shape[0], 10, replace=False)

for i, idx in enumerate(indices):
    signal = X[idx]
    label = np.argmax(Y[idx])

    np.save(f"signal_{i}.npy", signal)
    
    # save label also
    with open(f"signal_{i}_label.txt", "w") as f:
        f.write(str(label))

print("✅Signals + labels saved")
```
